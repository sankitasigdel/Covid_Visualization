```python
unique_country = data_full["Country/Region"].unique()
Dataframe = {}
Sub_DF = []
for x,i in enumerate(unique_country):
    Dataframe.update({f"df_{i}" : data_full[data_full["Country/Region"] == i]})
    Sub_DF.append(f"df_{i}")
Dataframe[Sub_DF[-1]][-10:-1]
```

```python
x1 = Dataframe[Sub_DF[-1]]["Date"][-10:-1]
y1 = Dataframe[Sub_DF[-1]]["Confirmed"][-10:-1]
y2 = Dataframe[Sub_DF[-1]]["Deaths"][-10:-1]
y3 = Dataframe[Sub_DF[-1]]["Recovered"][-10:-1]
y4 = Dataframe[Sub_DF[-1]]["Active"][-10:-1]
country_name = Dataframe[Sub_DF[-1]]["Country/Region"].unique()[0]

plt.plot(x1, y1, label='Cases', marker = "*", color = "green")
plt.plot(x1, y2, label='Deaths',marker = "o", color = "red")
plt.plot(x1, y3, label='Recovered', marker = "*" ,color = "orange")
plt.plot(x1, y4, label='Active',marker = "o", color = "blue")
  
plt.xlabel('Time')
plt.ylabel('People Count')
plt.title(f"Cases, Deaths, Recovered in {country_name}")
plt.xticks(rotation=90)
  
plt.legend()
plt.show()
```
